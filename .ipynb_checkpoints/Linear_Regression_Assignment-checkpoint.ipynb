## Import Libraries

```python
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Our Goal: to predict the price given a set of attributes.

## Load the Data

```python
from sklearn import datasets
boston = datasets.load_boston()
```

```python
X_boston,y_boston = boston.data, boston.target
print('Shape of data:', X_boston.shape, y_boston.shape)
```

```python
print('Keys:', boston.keys())
print('Feature names:',boston.feature_names)
```

## EDA (Exploratory Data Analysis)

## Q1: Create a dataframe and Save that dataset inside it.

```python
#new dataframe
df_boston = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
#adding "Price" column
df_boston['Price']=boston.target
```

## Q2: Print the head rows of the dataframe.

```python
df_boston.head()
```

## Q3: Use histogram to show the distribution of House Prices.

```python
#displaying histogram
sns.histplot(data=df_boston, x='Price', bins=10, element="step");
```

## Q4: Use a heatmap to show the correlation between features and the target labels.

```python
#setting the heatmap size
sns.set(rc={"figure.figsize":(15, 10)})
#showing the heatmap
sns.heatmap(df_boston.corr(), annot=True);
```

## Q5: Use a lmplot to draw the relations between price and LSTAT.

```python
#setting the figure's style
sns.set_style("white")

#showing implot and enhance its size
sns.lmplot(x ='Price', y ='LSTAT', data = df_boston, height=8, aspect=2);
```

## Q6: Use a lmplot to draw the relations between price and RM.

```python
#showing implot and enhance its size
sns.lmplot(x ='Price', y ='RM', data = df_boston, height=8, aspect=2);
```

```python
df_boston.columns
```

## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
from sklearn.model_selection import train_test_split
```

```python
#setting x and y values
X = df_boston.drop(['Price'], axis=1)
y = df_boston["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
```

```python
lrm = LinearRegression()
```

## Q9: Train the Model.

```python
lrm.fit(X_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
#showing the intercept
print(lrm.intercept_) 
```

```python
#setting and showing the coefficient
coeff_df = pd.DataFrame(lrm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
#y predictions
predictions = lrm.predict(X_test)
predictions
```

```python
#y real values
Real_Values = np.array(y_test)
Real_Values
```

```python
#setting figure size
sns.set(rc={"figure.figsize":(7, 5)})
#setting figure style
sns.set_style("white")
#showing real values and predicitions scatter
plt.scatter(Real_Values,predictions);
```

## Residual Histogram

```python
#showing the histogram of real values and predictions
sns.distplot(( Real_Values - predictions), bins=20);
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(Real_Values, predictions))
print('MSE:', metrics.mean_squared_error(Real_Values, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Real_Values, predictions)))
```

```python
#importing r2 score
from sklearn.metrics import r2_score
#printing the value
print('R^2:', r2_score(Real_Values, predictions))
```
